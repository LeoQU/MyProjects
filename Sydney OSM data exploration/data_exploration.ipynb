```python
import re
import pandas as pd

nodes = pd.read_csv('nodes.csv')
nodes_tags = pd.read_csv('nodes_tags.csv')

ways = pd.read_csv('ways.csv')
ways_tags = pd.read_csv('ways_tags.csv')
ways_nodes = pd.read_csv('ways_nodes.csv')
```

```python
# regular expression checking for each column in nodes.csv and ways.csv files

def re_check(df):
    for column_name in df.columns.values:
        count = 0
        print ''
        print "Checking %s..." % column_name
        for row in range(1000):
            if not eval(column_name + '_re').match( str(df.loc[row, column_name]) ):
                print df.loc[row, column_name]
                count += 1
        print "Number of Unexpected Data:", count

# regular expressions for nodes
id_re = re.compile( r'\d{6}' )
lat_re = re.compile( r'-?\d{1,3}\.\d{3,7}' )
lon_re = re.compile( r'-?\d{1,3}\.\d{3,7}' )
user_re = re.compile( r'[^=\+/&<>;\'"\?%#$@\,\. \t\r\n]' )
# user_re = re.compile( r'[a-zA-Z0-9]' )
uid_re = re.compile( r'\d{3,7}' )
version_re = re.compile( r'\d{1}' )
changeset_re = re.compile( r'\d{4,8}' )
timestamp_re = re.compile( r'20[0-1][0-9]-[0-1]\d-[0-3]\dT[0-2]\d:[0-5]\d:[0-5]\dZ' )

print 'Nodes Data:'
re_check(nodes)
print ''

print 'Ways Data:'
re_check(ways)
print ''
```

```python
# check suburb names in nodes_tags.csv

with open('Sydney_suburbs.csv', 'r') as f:
    c = f.readline()
    Sydney_suburbs_list = c.split(' - ')

for i in nodes_tags[ (nodes_tags['type'] == 'addr') & (nodes_tags['key'] == 'city') ].groupby('value').groups.keys():
    if i not in Sydney_suburbs_list:
        print i
```

```python
# check postcodes in nodes_tags.csv

postcode_list = list( nodes_tags[ (nodes_tags['type'] == 'addr') & (nodes_tags['key'] == 'postcode') ]['value'] )
postcode_re = re.compile( r'2\d{3}' )
[postcode for postcode in postcode_list if not postcode_re.match(postcode)]
```

```python
# check address names in nodes_tags.csv

street_list = nodes_tags[ (nodes_tags['type'] == 'addr') & (nodes_tags['key'] == 'street') ]['value']

street_name_re = re.compile( r'(?:([A-Za-z0-9- \']+ (?:Street|Road|Avenue|Plaza|Way|Place|Parkway|Parade|Grove|Market|Close|Lane|Highway|Row|Esplanade|Drive|Boulevard|Crescent|Arcade|Circuit|Court|Terrace|Square))|Broadway|Kingsway|The Piazza|The Promanade|Victoria Adventure|Chatswood Mall)')

for street in street_list:
    if not street_name_re.match( street ):
        print street
```
